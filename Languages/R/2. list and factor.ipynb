# 5. 리스트와 팩터


## 1. 리스트
서로 다른 자료형의 값들을 1차원 배열에 저장하고 다룰 수 있도록 해주는 수단

```r
ds <- c(90, 85, 70, 84)
my.info <- list(name='Tom', age=60, status=TRUE, score=ds)

my.info # 리스트에 저장된 내용 모두 출력
my.info[1] # key-value 모두 출력                       # $name = 'Tom'
my.info[[1]] # 리스트의 첫번째 value를 출력               # "Tom"
my.info$name # 리스트에서 key가 name인 값 출력            # "Tom"
my.info[4] # 리스트의 네번째 key-value 모두               # $score = 90 85 70 84
my.info[[4]] # # 리스트의 네번째 value 출력               # 90 85 70 84
```

## 2. 팩터 (벡터의 일종)
- 문자형 데이터가 저장된 벡터의 일종
- 예를 들어 성별, 혈액형, 선호정당등과 같이 **저장할 문자값들이 몇 종류로 정해져있을때 팩터 사용**

```r
bt <- c('A', 'B', 'B', 'O', 'AB', 'A') # 문자형 벡터 bt 정의
bt.new <- factor(bt) # 팩터 bt.new 정의

bt # 벡터 bt 내용 출력                       # 'A''AB''B''O'
bt.new # 팩터 bt.new 내용 출력               # Levels: 'A''AB''B''O'

bt.new[5] # 팩터의 다섯번째 값 출력            # "AB"
levels(bt.new) # 팩터에 저장된 value 종류 출력 # 'A''AB''B''O'
```
