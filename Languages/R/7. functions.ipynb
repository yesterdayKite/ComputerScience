# 2. user defined definition

> 함수명 <- function(매개변수 목록) {
    실행할 명령문(들)
    return (함수의 실행결과)
}

함수 기본 정의

```r
mymax <- function(x,y) {
  num.max <- x
  if (y > x) [
    num.max <- y
  ]
  return (num.max)
}

mymax(10, 15)
```

매개변수에 초기값 설정하기

```r
mydiv <- function(x, y = 2){
  result <- x/y
  return (result)
}

mydiv(x=10, y=3)
mydiv(10,3)
mydiv(10) #x값만 입력되어 y값은 default값이 들어감
```

리턴값이 여러개일 경우

```r
myfunc <- function(x,y) {
  val.sum <- x+y
  val.mul <- x*y
  return (list(sum=val.sum , mul=val.mul))
}

result <- myfunc(5,8)

s <- result$sum #결과리스트에서 5,8의 합
m <- result$mul #결과리스트에서 5,8이 곱

```

# 3. which 

조건에 맞는 데이터 위치 찾기

- 데이터 분석을 하다보면 자신이 원하는 데이터가 벡터나 매트릭스, 데이터 프레임 안에서 어디에 위치하고 있는지를 알기 원하는 때가 있음
- 예를 들어, 50명의 학생 성적이 저장된 벡터가 있는데 가장 성적이 좋은 학생은 몇 번째에 있는지를 알고 싶은 경우
- 이런 경우 편리하게 사용할 수 있는 함수가 which(), which.max(), which.min() 함수

```r
score <- c(76, 84, 69, 50, 95, 60, 82, 71, 88, 84)

which(score==69) # 69점인 학생 인덱스 리턴       # 3
which(score>=85) # 85점 이상인 학생 인덱스 리턴   # 5 9

max(score) #최고 점수 value 
which.max(score) # 최고점수의 index

```

```r
score <- c(76, 84, 69, 50, 95, 60, 82, 71, 88, 84)

idx <- which(score<=60) # 60점 이하인 값들의 index
score[idx] <- 61       # 60점 이하인 값들 일괄적으로 61점으로 overwrite


idx <- which(score>=80) # 80점 이상인 값들의 index
score.high <- score[idx] # 성적이 80이상인 값들만 추출하여 저장
score.high               # 값들 출력
```

```r
idx <- which(iris$Petal.Length > 5.0) #꽃잎의 길이가 5.0 이상인 값들의 인덱스
idx

iris.big <- iris[idx,] # 인덱스에 해당하는 값만 추출하여 저장
iris.big
```

```r
idx <- which(iris[,1:4]>5.0, arr.ind=TRUE) # 1~4열의 값들 중 5.0보다 큰 값의 행과 열의 위치
idx
```
