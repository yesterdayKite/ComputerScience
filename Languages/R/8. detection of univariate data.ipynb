# 1. 자료의 종류

#### 자료의 특성에 따른 분류
- 범주형자료 categorical data
  - 질적자료 (qualitative data)
  - 범주 / 그룹으로 구분할 수 있는 값으로 구성된 자료
  - 기본적으로 숫자로 표현할 수 없음 (=대소연산 불가능)
    - 성별을 0,1처럼 숫자로 표기했다고해서 계산 가능한 연속형 자료가 되는것은 아님
  - ex) 성별(M,F), 혈액형(A,B,O,AB)
- 연속형자료 numerical data
  - 양적자료 (quantitative data)
  - 대소 비교, 평균, 최댓값등 연산이 가능함


#### 변수 갯수에 따른 분류
- 단일 변수 자료 univariate data
  - 일변량 자료
  - 하나의 변수로만 이루어져있음 (몸무게 테이블)
  - R : 벡터에 저장
- 다중 변수 자료 multivariate data
  - 다변량 자료
  - 두개 이상의 변수로 구성된 자료
  - ex) 키,몸무게,성별 테이블
  - R : 매트릭스 / 데이터 프레임에 저장
    - 하나의 열 column이 하나의 변수를 나타냄
    - column 의 갯수 = 변수의 갯수


## 2. 단일변수 범주형 자료의 탐색 

- 단일변수 범주형 자료
  - = 일변량 질적 자료
  - 특성이 하나 && 자료의 특성이 범주형임
  - 막대/원형그래프등 가능


도수분포표 작성

```r
# 단일변수 범주형 자료
favorite <- c('WINTER', 'SUMMER', 'SPRING', 'SUMMER', 'SUMMER',
              'FALL', 'FALL', 'SUMMER', 'SPRING', 'SPRING')

table(favorite) # 도수 분포표 계산
table(favorite)/length(favorite) # 비율 출력
```

막대그래프 작성

```r
ds <- table(favorite)
ds
barplot(ds, main='favorite season')
```

원그래프 작성

```r
ds <- table(favorite)
ds
pie(ds, main='favorite season')
```

숫자로 표현된 범주형 자료



```r
# (1=초록, 2=빨강, 3=파랑으로 나타냄)
favorite.color <- c(2,3,2,1,1,2,2,1,3,2,1,3,2,1,2)
ds <- table(favorite.color) # 도수분포표로 변환

barplot(ds, main='favorite color') # 막대그래프
colors <- c('green', 'red', 'blue')
names(ds) <- colors # 자료값 1,2,3 을 green, red, blue로 변경

barplot(ds, main='favorite color') # 막대그래프
pie(ds, main='favorite color', col=colors)
```

# 3. 단일변수 연속형 자료의 탐색

- 연속형 자료는 관측값들이 크기를 가지기 때문에 다양한 분석 가능
  - 평균 
  - 중앙값 (median)
  - 절사평균 (trimmed mean)



```r
weight <- c(60, 62, 64, 65, 78, 69)

mean(weight) #평균
median(weight) #중앙값
mean(weight, trim=0.2) #절사평균(상하위, 20% 제외)
mean(weight.heavy, trim=0.2) #절사평균(상하위, 20% 제외)
```

4분위수
image.png

```r
mydata <- c(60, 62, 64, 65, 68, 69, 120)
quantile(mydata) #4분위수
quantile(mydata, (0:10)/10) # 10%단위로 구간을 나누어 계산
summary(mydata)
```

분산과표준편차

```r
mydata <- c(60, 62, 64, 65, 68, 69, 120)
var(mydata)          # 분산
sd(mydata)           # 표준편차
range(mydata)        # 값의 범위
diff(range(mydata))  #최댓값, 최솟값의 차이
```
