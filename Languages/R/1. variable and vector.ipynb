```r
###################
###### 연속적 숫자로 이루어진 벡터
###################

v1 <- 50:90 # 50이상 90이하를 벡터로 만듦
v2 <- c(1,2,5, 50:90) # 1,2,5와 50~90 의 숫자들로 벡터 만듦

v2 # 출력
```

```r
###################
###### 일정한 간격의 숫자로 이루어진 벡터
###################

v3 <- seq(1, 101, 3) # 1부터 101까지 3단위로 
v3
```

```r
###################
###### 반복된 숫자로 이루어진 벡터
###################

v5<- rep(1, times=5) # 1을 5번 반복
v6 <- rep(1:5, times=3) # 1에서 5까지 3번 반복
v7 <- rep(c(1,5,9), times=3) # 1,5,9를 3번 반복

v7
```

```r
###################
###### 벡터의 원소값에 이름 지정 : names
###################
score <- c(90, 85, 70)
names(score) # score에 저장된 값들의 이름을 보이시오 # NULL
names(score) <- c("John", "Tom", "Jane")
names(score) # score에 저장된 값들의 이름을 보이시오 # Joh, Tom, Jane
score # 이름과 값이 함께 출력
```

```r
###################
###### 벡터에서 여러개 값을 한번에 추출
###################
d <- c(1,4,3,7,8)
d[c(1,3,5)] # 1, 3, 5번재 값 출력
d[1:3] # 1~3번째 값 출력
d[seq(1,5,2)] # 홀수번째 값 출력
d[-2] # 2번째 값 제외하고 출력
d[-c(3:5)] # 3~5번째 값은 제외하고 출력
```

```r
###################
###### 벡터에서 이름으로 값 추출하기
###################

GNP <- c(2090, 2450, 960)
names(GNP) <- c("Korea", "Japan", "Nepal")
GNP[1] #2090
GNP["Korea"] # 2090
GNP["Korea", "Nepal"]
```

# 3. vector operation

- sum() : 벡터에 포함된 값들의 합
- mean() : 값들의 평균 
- median() : 값들의 중앙값
- max(), min() : 값들의 최댓값, 최솟값
- var() : 값들의 분산
- sd() : 값들의 표준편차
- sort() : 값들을 정렬 (기본 오름차순)
- range() : 값들의 범위 (최솟값~최댓값)
- length() : 포함된 값들의 갯수 (길이)

```r
###########################################################
# 벡터에 대한 산술연산은 벡터 안에 포함된 값들에 대한 연산으로 바뀌어 실행
d <- c(1,4,3,7,8)
2*d # 각 원소에2를 곱한값 출력


###########################################################
# 벡터간의 대응되는 위치에 있는 값끼리의 연산으로 바꾸어 연산
x<-c(1,2,3)
y<-c(4,5,6)
x+y # 각 원소를 더함
z<- x*y # 각원소 곱한 값으로 새로운 벡터 생성
```
