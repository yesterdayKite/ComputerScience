# 2. 상관 분석

- 상관분석
  - correlation analysis
  - 얼마나 선형성을 보이는지 수치상으로 나타낼 수 있는 방법
  - Pearson's correlation coerricient 
    - 피어슨 상관계수
    - image.png
    - -1 <= r <= 1
    - r>0 : 양의 상관관계
    - r<0 : 음의 상관관계
    - r이 1이나 -1에 가까울 수록 x,y의 상관성이 높음

R을 이용한 상관계수의 계산
image.png

```r

beers <-  c(5,2,9,8,3,7,3,5,3,5)
bal <- c(0.1,0.03,0.19,0.12,0.04,0.0095,0.07,0.06,0.02,0.05)

tbl <- data.frame(beers, bal) #데이터 프레임 생성
tbl
```

```r
plot(bal~beers, data=tbl) # 산점도 그리기 

res <- lm(bal~beers, data=tbl) #회귀식 도출
abline(res) #회귀 선 그리기
cor (beers, bal) # 상관계수 계산
```

```r
# iris 데이터로 상관계수 계산
cor(iris[,1:4]) # 4개 변수간 상관성 분석
```
