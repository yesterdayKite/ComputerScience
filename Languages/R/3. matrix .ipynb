### matrix

- 2차원 테이블 형태 자료구조 
- 모든 셀에 저장되는 값은 동일한 자료형이어야함

```r
#############################
##### 기본 매트릭스 만들기 
#############################


# arg0 = 매트릭스에 저장될 값 (1:20)
# arg1 = 행의 수 (nrow=4)
# arg2 = 열의 수 (ncol=5)
z <- matrix(1:20, nrow=4, ncol=5) # default = 열방향으로 채우기
z
```

```r
#############################
##### 매트릭스에 저장될 값들을 행 방향으로 채우기
#############################

z2 <- matrix(1:20, nrow=4, ncol=5, byrow=T)  # 행방향으로 채우기
z2
```

```r
#############################
## 기존 매트릭스에 벡터 추가하여 새로운 매트릭스 만들기
#############################

x <- 1:4  #벡터 x 생성
y <- 5:8  #벡터 y 생성
z <- matrix(1:20, nrow=4, ncol=5) # 매트릭스 z 생성

m1 <- cbind(x,y) # x와 y를 열방향으로 결합하여 매트릭스 생성
m1

m2 <- rbind(x,y) # x와 y를 행방향으로 결합하여 매트릭스 생성
m2

m3 <- rbind(m2,x) # 매트릭스 m2와 벡터 x를 행방향으로 결합
m3

m4 <- cbind(z,x) # 매트릭스 z와 벡터x를 열방향으로 결합
m4
```

#### 매트릭스에서의 값 추출 

```r
#############################
###### 인덱스 사용

z <- matrix(1:20, nrow=4, ncol=5) # 매트릭스 z 생성
z

z[2,3]  # 2행 3열
z[1,4]  # 1행 4열

z[2,]   # 2행에 있는 모든 값
z[,4]   # 4열에 있는 모든 값 




#############################
###### 여러개값 동시 추출

z[2, 1:3] # 2행중 1~3열 값
z[1, c(1,2,4)] # 1행중 1,2,4열 값
z[1:2, ] # 1,2행에 있는 모든 값
z[,c(1,4)] # 1,4열에 있는 모든 값

```

#### 매트릭스의 행과 열에 이름 지정

```r
score <- matrix(c(90, 85, 69, 78,
                  85, 96, 49, 95,
                  90, 80, 70, 60),
                  nrow=4, ncol=3)

rownames(score) <- c('John', 'Tom', 'Mark', 'Jane') #행에 이름 지정
colnames(score) <- c('English', 'Math', 'Science')  #열에 이름 지정

score
```

#### 행과 열에 지정한 이름을 이용하여 매트릭스 값 추출

```r
score['John', 'Math'] #John행의 Math열 value

score('Tom', c('Math', 'Science')) # Tom행의 Math, Science열 value

score('Mark') #Mark행의 모든 열 (math, science, english)

score(, 'English') # 모든 학생의 영어 성적

rownames(score)    # 모든 행의 이름
colnames(score)    # 모든 열의 이름
colnames(score)[2] #열의 이름 중 두번째 값
```
